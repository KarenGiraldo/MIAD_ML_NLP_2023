### Punto 1 - Árbol de decisión manual

En la celda 1 creen un árbol de decisión **manualmente**  que considere los set de entrenamiento y test definidos anteriormente y presenten el RMSE y MAE del modelo en el set de test.

```python
X_train_array = X_train.to_numpy()
y_train_array = y_train.to_numpy()
X_test_array = X_test.to_numpy()

class NodoArbolDecision:
    total_nodos = 0  

    def __init__(self, profundidad=0, profundidad_max=10, numero_nodo=0):
        self.profundidad = profundidad
        self.profundidad_max = profundidad_max
        self.numero_nodo = numero_nodo
        self.es_hoja = False
        self.caracteristica_division = None
        self.umbral = None
        self.izquierda = None
        self.derecha = None
        self.prediccion = None
        NodoArbolDecision.total_nodos += 1  # Incrementar el contador de nodos cada vez que se crea un nodo

    def ajustar(self, X, y):
        if self.profundidad == self.profundidad_max or len(X) < 2:
            self.es_hoja = True
            self.prediccion = np.mean(y)
            return

        mejor_caracteristica, mejor_umbral = self.encontrar_mejor_division(X, y)

        if mejor_caracteristica is None:
            self.es_hoja = True
            self.prediccion = np.mean(y)
            return

        self.caracteristica_division = mejor_caracteristica
        self.umbral = mejor_umbral

        indices_izquierda = X[:, mejor_caracteristica] < mejor_umbral
        indices_derecha = X[:, mejor_caracteristica] >= mejor_umbral

        self.izquierda = NodoArbolDecision(self.profundidad + 1, self.profundidad_max, self.numero_nodo * 2 + 1)
        self.derecha = NodoArbolDecision(self.profundidad + 1, self.profundidad_max, self.numero_nodo * 2 + 2)

        self.izquierda.ajustar(X[indices_izquierda], y[indices_izquierda])
        self.derecha.ajustar(X[indices_derecha], y[indices_derecha])

    def encontrar_mejor_division(self, X, y):
        mejor_mse = float('inf')
        mejor_caracteristica = None
        mejor_umbral = None

        for indice_caracteristica in range(X.shape[1]):
            umbrales = np.unique(X[:, indice_caracteristica])
            for umbral in umbrales:
                indices_izquierda = X[:, indice_caracteristica] < umbral
                indices_derecha = X[:, indice_caracteristica] >= umbral

                if len(y[indices_izquierda]) < 1 or len(y[indices_derecha]) < 1:
                    continue

                mse = self.calcular_mse(y[indices_izquierda], y[indices_derecha])

                if mse < mejor_mse:
                    mejor_mse = mse
                    mejor_caracteristica = indice_caracteristica
                    mejor_umbral = umbral

        return mejor_caracteristica, mejor_umbral

    def calcular_mse(self, y_izquierda, y_derecha):
        media_izquierda = np.mean(y_izquierda)
        media_derecha = np.mean(y_derecha)
        mse_izquierda = np.mean((y_izquierda - media_izquierda) ** 2)
        mse_derecha = np.mean((y_derecha - media_derecha) ** 2)

        return (mse_izquierda * len(y_izquierda) + mse_derecha * len(y_derecha)) / (len(y_izquierda) + len(y_derecha))

    def predecir(self, X):
        if self.es_hoja:
            return np.array([self.prediccion] * len(X))

        indices = X[:, self.caracteristica_division] < self.umbral
        predicciones = np.empty(len(X))
        predicciones[indices] = self.izquierda.predecir(X[indices])
        predicciones[~indices] = self.derecha.predecir(X[~indices])

        return predicciones
# Ejecutar ajuste y predicción
arbol = NodoArbolDecision(profundidad_max=5)
arbol.ajustar(X_train_array, y_train_array)
y_pred_arbol_m = arbol.predecir(X_test_array)

# Cálculo de RMSE y MAE
rmse_arbol_decision = np.sqrt(mean_squared_error(y_test, y_pred_arbol_m))
mae_arbol_decision = mean_absolute_error(y_test, y_pred_arbol_m)

print(f"Total de nodos en el árbol: {NodoArbolDecision.total_nodos}")
print(f"RMSE: {rmse_arbol_decision}, MAE: {mae_arbol_decision}")
```

El árbol de decisión tiene un total de 63 nodos. Esto incluye tanto nodos hoja (nodos terminales donde se realizan predicciones) como nodos internos (nodos donde se toman decisiones basadas en características).

En cuanto al error cuadratico medio (rmse) dió un valor de 1.721  lo que nos puede decir que en promedio las predicciones se desvian 1721 del valor real.

### Punto 2 - Bagging manual

En la celda 2 creen un modelo bagging **manualmente** con 10 árboles de regresión y comenten sobre el desempeño del modelo.

```python
# Celda 2
def entrenar_modelo_bagging(X_train, y_train, n_estimators=10):
    n_samples = X_train.shape[0]
    modelos = []
    for _ in range(n_estimators):
        X_sample, y_sample = resample(X_train, y_train, n_samples=n_samples)
        modelo = DecisionTreeRegressor()
        modelo.fit(X_sample, y_sample)
        modelos.append(modelo)
    return modelos

def predecir_modelo_bagging(modelos, X):
    predicciones = np.array([modelo.predict(X) for modelo in modelos])
    return np.mean(predicciones, axis=0)
```

```python
# Entrenamiento del modelo de bagging
modelos_bagging = entrenar_modelo_bagging(X_train_array, y_train_array, n_estimators=10)

# Predicción con el modelo de bagging
y_pred_bagging = predecir_modelo_bagging(modelos_bagging, X_test_array)

rmse_bagging = np.sqrt(mean_squared_error(y_test, y_pred_bagging))
mae_bagging = mean_absolute_error(y_test, y_pred_bagging)

print(f"RMSE del modelo de bagging: {rmse_bagging}")
print(f"MAE del modelo de bagging: {mae_bagging}")
```

Se realizó la cconsutrccion del modelo baggin utilizando arboles de de decisión de la libreria sklearn

Hubo un aumento en RMSE y MAE:
El RMSE ha aumentado de 1.721 a 1.804
El MAE ha aumentado de 1.292 a 1.346

Al utilizar 10 arboles puede que el valor no sea optimo 

### Punto 3 - Bagging con librería

En la celda 3, con la librería sklearn, entrenen un modelo bagging con 10 árboles de regresión y el parámetro `max_features` igual a `log(n_features)` y comenten sobre el desempeño del modelo.

```python
# Celda 3
from sklearn.ensemble import BaggingRegressor

#número de características a considerar según el logaritmo del número de características
max_caracteristicas = int(np.log(X_train.shape[1]))

# Creación del modelo de Bagging
modelo_bagging = BaggingRegressor(
    base_estimator=DecisionTreeRegressor(),
    n_estimators=10,
    max_features=max_caracteristicas,
    random_state=42
)

modelo_bagging.fit(X_train, y_train)

y_pred_bagging_2 = modelo_bagging.predict(X_test)

rmse_bagging_2 = np.sqrt(mean_squared_error(y_test, y_pred_bagging_2))
mae_bagging_2 = mean_absolute_error(y_test, y_pred_bagging_2)

print(f"RMSE: {rmse_bagging_2:.2f}")
print(f"MAE: {mae_bagging_2:.2f}")
```

Comparado con el modelo de bagging manual sin especificar max_features (RMSE: 1.804, MAE:1.346), observamos que el modelo actual con max_features=log(n_features) ha mostrado un aumento en ambos errores (RMSE y MAE). Este incremento sugiere que la restricción impuesta por max_features puede no estar beneficiando al modelo en este contexto particular.

Un RMSE y MAE elevados pueden indicar un subajuste.

### Punto 4 - Random forest con librería

En la celda 4, usando la librería sklearn entrenen un modelo de Randon Forest para regresión  y comenten sobre el desempeño del modelo.

```python
# Celda 4
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor(n_estimators=100, random_state=42)

forest_model.fit(X_train, y_train)

# Realización de predicciones sobre el conjunto de prueba
y_pred_random = forest_model.predict(X_test)

# Cálculo de RMSE y MAE
rmse_radom = np.sqrt(mean_squared_error(y_test, y_pred_random))
mae_random = mean_absolute_error(y_test, y_pred_random)

print(f"RMSE: {rmse_radom:.2f}")
print(f"MAE: {mae_random:.2f}")
```

El Random Forest ha mejorado ligeramente en términos de RMSE y ha mostrado un rendimiento similar en MAE. Esto sugiere que la agregación y diversificación en Random Forest están contribuyendo positivamente al manejo de la varianza y la mejora de la precisión general.

El Random Forest ha mostrado una mejora significativa tanto en RMSE como en MAE. Este resultado reafirma la eficacia de Random Forest en comparación con un bagging más restringido, subrayando la importancia de la elección adecuada de parámetros en técnicas de ensamble.

Reducción de Varianza: Al combinar múltiples árboles, Random Forest reduce la varianza sin incrementar el sesgo.

### Punto 5 - Calibración de parámetros Random forest

En la celda 5, calibren los parámetros max_depth, max_features y n_estimators del modelo de Randon Forest para regresión, comenten sobre el desempeño del modelo y describan cómo cada parámetro afecta el desempeño del modelo.

```python
# Celda 5
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, make_scorer
param_grid = {
    'max_depth': [10, 20, 30, None],  # Profundidades del árbol
    'max_features': ['auto', 'sqrt', 'log2'],  # Número de características a considerar para las mejores divisiones
    'n_estimators': [50, 100, 200]  # Número de árboles en el bosque
}

# Modelo base para tunear
rf = RandomForestRegressor(random_state=42)

# Creación del GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, scoring='neg_mean_squared_error', verbose=2)
grid_search.fit(X_train, y_train)
```

```python
# Mejores parámetros y modelo
mejor_rf = grid_search.best_estimator_
print("Mejores parámetros: ", grid_search.best_params_)

# Predicción y cálculo del RMSE
y_pred_random_cal = mejor_rf.predict(X_test)
rmse_random_cal = np.sqrt(mean_squared_error(y_test, y_pred_random_cal))
mae_random_cal = mean_absolute_error(y_test, y_pred_random_cal)
print(f"RMSE optimizado: {rmse_random_cal:.2f}")
print(f"MAE optimizado: {mae_random_cal:.2f}")
```

max_depth: 10 los árboles dentro del bosque no deben tener una profundidad superior a 10.

max_features: 'sqrt' es la raíz cuadrada del número total de características para cada división 

n_estimators: 200 - El modelo utiliza 200 árboles, lo cual es una cantidad considerable para asegurar que el modelo es estable.

Hay una mejora notable en ambos RMSE y MAE respecto a los modelos previos sin calibración y otros modelos de ensamble. Esto indica una alta eficacia del Random Forest calibrado en capturar y predecir correctamente las variaciones en los datos de prueba

### Punto 6 - XGBoost con librería

En la celda 6 implementen un modelo XGBoost de regresión con la librería sklearn y comenten sobre el desempeño del modelo.

```python
# Celda 6
from xgboost import XGBRegressor

modelo_xgb = XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

modelo_xgb.fit(X_train, y_train)

y_pred_xgb = modelo_xgb.predict(X_test)

# Cálculo de RMSE y MAE
rmse_xgb = np.sqrt(mean_squared_error(y_test, y_pred_xgb))
mae_xgb = mean_absolute_error(y_test, y_pred_xgb)

print(f"RMSE: {rmse_xgb:.2f}")
print(f"MAE: {mae_xgb:.2f}")
```

Comparado con los modelos anteriores (Random Forest y Bagging con ajustes específicos), el modelo XGBoost muestra un mejoramiento notable en el RMSE y el MAE:
Random Forest (Calibrado): RMSE de 1.564y MAE de 1.147
Modelo de Bagging (Mejorado): RMSE y MAE mayores que los del modelo XGBoost.
Estos resultados sugieren que XGBoost, con su capacidad para manejar el sobreajuste.

### Punto 7 - Calibración de parámetros XGBoost

En la celda 7 calibren los parámetros learning rate, gamma y colsample_bytree del modelo XGBoost para regresión, comenten sobre el desempeño del modelo y describan cómo cada parámetro afecta el desempeño del modelo.

```python
# Celda 7
param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],  # Tasa de aprendizaje
    'gamma': [0, 0.1, 0.3],            # Parámetro de regularización
    'colsample_bytree': [0.3, 0.7, 1.0] # Porcentaje de características usadas por árbol
}

# Modelo base de XGBoost
modelo_xgb_2 = XGBRegressor(n_estimators=100, max_depth=3, random_state=42)

# Scorer para optimizar el menor RMSE
scorer = make_scorer(mean_squared_error, greater_is_better=False)

# Configuración del GridSearchCV
grid_search = GridSearchCV(estimator=modelo_xgb_2, param_grid=param_grid, cv=3, scoring=scorer, verbose=2)
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
best_xgb = grid_search.best_estimator_

y_pred_xgb_2 = best_xgb.predict(X_test)

# Cálculo de RMSE y MAE
rmse_xgb_2 = np.sqrt(mean_squared_error(y_test, y_pred_xgb_2))
mae_xgb_2 = mean_absolute_error(y_test, y_pred_xgb_2)

print("Mejores parámetros:", best_params)
print(f"RMSE Optimizado: {rmse_xgb_2:.2f}")
print(f"MAE Optimizado: {mae_xgb_2:.2f}")
```

* colsample_bytree: 0.7  cada árbol en el ensamble usa el 70% de las características disponibles al hacer divisiones. 
* gamma: 0 gamma de 0 sugiere que no hay una reducción mínima de pérdida requerida para hacer divisiones adicionales en un árbol. 
* learning_rate: 0.1 - Una tasa de aprendizaje de 0.1 

Comparando estos resultados con evaluaciones previas:

Modelo de Random Forest Calibrado:
RMSE: 1.564
MAE: 1.147
Modelo de XGBoost sin Calibrar:
RMSE: 1.542
MAE: 1.130
El modelo XGBoost calibrado muestra una mejora en el RMSE en comparación con ambas configuraciones anteriores, y una ligera mejora en el MAE comparado con el modelo de Random Forest

### Punto 8 - Comparación y análisis de resultados
En la celda 8 comparen los resultados obtenidos de los diferentes modelos (random forest y XGBoost) y comenten las ventajas del mejor modelo y las desventajas del modelo con el menor desempeño.

```python
# Celda 8
def compararModelos (models, data, score): 
    better = min(data)
    colores = ['#87CEEB' if value != better else '#6B8E23' for value in data]
    fig, ax = plt.subplots(figsize=(12,5))
    bar = ax.bar(models,data,linewidth=0.8, color=colores, edgecolor="silver")
    ax.set_title("Comparación del desempeño de los modelos")
    ax.set_xlabel("Descripción del modelo")
    ax.set_ylabel(score)
    bar_labels=[f'{value:,.0f}' for value in data]
    ax.bar_label(bar, bar_labels, label_type='center', padding=20)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.set_facecolor("whitesmoke")
    ax.spines["left"].set_color("gray")
    ax.spines["bottom"].set_color("gray")
    ax.tick_params(axis='x', labelsize=8)
    #ax.set_xticklabels(models, rotation=90)
    plt.gca().yaxis.set_major_formatter('{:,.0f}'.format)
    plt.tight_layout()
    plt.show()

modelos = ["Arbol de decisión", "Bagging","Bagging función","Random Forest","Random Forest Calibrado",
           "XGBoost","XGBoost Calibrado"]
rmses = [rmse_arbol_decision,rmse_bagging,rmse_bagging_2,rmse_radom,rmse_random_cal,
        rmse_xgb,rmse_xgb_2]

maes = [mae_arbol_decision,mae_bagging,mae_bagging_2,mae_random,mae_random_cal,
        mae_xgb,mae_xgb_2]

compararModelos(modelos, rmses, "RMSE")
compararModelos(modelos, maes, "MAE")
```
