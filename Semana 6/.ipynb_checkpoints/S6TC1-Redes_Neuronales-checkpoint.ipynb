## Datos predicción probabilidad de admisión a la universidad

En este taller se usará el conjunto de datos de admisiones a una universidad proveniente de la base de datos de Kaggle. Cada observación contiene la probabilidad de que un estudiante sea admitido por la universidad, dadas distintas variables predictoras como el puntaje del examén TOEFL y GRE, el promedio (GPA), entre otras. El objetivo es predecir la probabilidad de admissión de cada estudiante. Para más detalles pueden visitar el siguiente enlace: [datos](https://www.kaggle.com/mohansacharya/graduate-admissions).

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Importación librerías
import numpy as np
#import keras 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
from keras import initializers
from keras import optimizers
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
#!pip install tensorflow
```

```python
# Carga de datos de archivo .csv
data = pd.read_csv('https://raw.githubusercontent.com/albahnsen/MIAD_ML_and_NLP/main/datasets/universityGraduateAdmissions.csv', index_col=0)
data.head()
```

```python
# Hitogramas de las varaibles del dataset
data.hist(bins=25, figsize=(15,15))
plt.show()
```

```python
# Definición de variables predictoras (X)
X = data.drop(data.columns[-1], axis=1)
# Definición de variable de interés (y)
Y = data[data.columns[-1]]
```

```python
# Separación de variables predictoras (X) y variable de interés (y) en set de entrenamiento y test usandola función train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(X,Y,test_size=0.3, random_state=22)
```

## Punto 1 - Red Neuronal de una capa

En la celda 1 creen una **red neuronal de una capa** con la librería Keras, que prediga la probabilidad de admisión de los estudiantes, usando los sets de entrenamiento y test definidos anteriormente. Pueden usar la función de perdida, el optimizador y el número de épocas que consideren pertinentes para el modelo, justificando su selección. Finalmente, grafiquen la pérdida del modelo vs el número de épocas en el set de entrenamiento y validación, y presenten el desempeño del modelo con las métricas error absoluto medio (MAE) y error cuadrático medio (MSE).

```python
# Celda 1
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.optimizers import Adam

scaler = StandardScaler()

xTrain_scaled = scaler.fit_transform(xTrain)
xTest_scaled = scaler.transform(xTest)

model = Sequential([
    Dense(1, input_dim=xTrain.shape[1], activation='sigmoid')
])

# Compilar el modelo
model.compile(optimizer=Adam(learning_rate=0.01), loss='mean_squared_error', metrics=['mean_absolute_error', 'mean_squared_error'])

# Entrenar el modelo
history = model.fit(xTrain_scaled, yTrain, validation_split=0.2, epochs=100, batch_size=10, verbose=1)
```

```python
import matplotlib.pyplot as plt

# Plot de la pérdida
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
# Evaluar el modelo
test_metrics_1 = model.evaluate(xTest_scaled, yTest, verbose=0)
print(f'Mean Absolute Error (MAE) on Test data: {test_metrics_1[1]}')
print(f'Mean Squared Error (MSE) on Test data: {test_metrics_1[2]}')
```

## Punto 2 - Red Neuronal multicapa

En la celda 2 creen una **red neuronal con dos capas** con la librería Keras, que prediga la probabilidad de admisión de los estudiantes usando los sets de entrenamiento y test definidos anteriormente. Pueden usar la función de perdida, el optimizador, el número de épocas y el número de neuronas que consideren pertinentes para el modelo, justificando su selección. Finalmente, grafiquen la pérdida del modelo vs el número de épocas en el set de entrenamiento y validación, y presenten el desempeño del modelo con las métricas error absoluto medio (MAE) y error cuadrático medio (MSE).

```python
# Celda 2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Número de neuronas en la capa oculta
hidden_neurons = 10  # Este número puede ajustarse según la complejidad necesaria

# Definición del modelo
model = Sequential([
    Dense(hidden_neurons, input_dim=xTrain.shape[1], activation='relu'),  # Capa oculta con activación ReLU
    Dense(1, activation='sigmoid')  # Capa de salida con activación sigmoidal para probabilidad
])

# Compilación del modelo
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error', 'mean_squared_error'])
```

```python
history = model.fit(xTrain_scaled, yTrain, validation_split=0.2, epochs=100, batch_size=10, verbose=1)
```

```python
import matplotlib.pyplot as plt

# Plot de la pérdida
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
# Evaluar el modelo en el conjunto de prueba
test_metrics_2 = model.evaluate(xTest_scaled, yTest, verbose=0)
print(f'Mean Absolute Error (MAE) on Test data: {test_metrics_2[1]}')
print(f'Mean Squared Error (MSE) on Test data: {test_metrics_2[2]}')
```

## Punto 3  - Red Neuronal multicapa

En la celda 3 creen **una red neuronal con más de una capa con la librería Keras, usando early stopping y dropout,** que prediga la probabilidad de admisión de los estudiantes con los sets de entrenamiento y test definidos anteriormente. Pueden usar la función de perdida, el optimizador, el número de épocas y el número de neuronas que consideren pertinentes para el modelo, justificando su selección. Finalmente, grafiquen la pérdida del modelo vs el número de épocas en el set de entrenamiento y validación, y presenten el desempeño del modelo con las métricas error absoluto medio (MAE) y error cuadrático medio (MSE).

```python
# Celda 3
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# Definición del modelo
model = Sequential([
    Dense(64, input_dim=xTrain.shape[1], activation='relu'),
    Dropout(0.5),  # 50% de Dropout
    Dense(32, activation='relu'),
    Dropout(0.5),  # 50% de Dropout
    Dense(1, activation='sigmoid')
])

# Compilación del modelo
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error', 'mean_squared_error'])

```

```python
# Configuración de Early Stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
# Entrenamiento del modelo
history = model.fit(xTrain_scaled, yTrain, validation_split=0.2, epochs=100, batch_size=10, verbose=1, callbacks=[early_stopping])
```

```python
import matplotlib.pyplot as plt

# Plot de la pérdida
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
# Evaluación del modelo en el conjunto de prueba
test_metrics_3 = model.evaluate(xTest_scaled, yTest, verbose=0)
print(f'Mean Absolute Error (MAE) on Test data: {test_metrics_3[1]}')
print(f'Mean Squared Error (MSE) on Test data: {test_metrics_3[2]}')
```

## Punto 4 - Comparación y análisis de resultados

En la celda 4 comparen los resultados obtenidos de las diferentes redes y comenten las ventajas del mejor modelo y las desventajas del modelo con el menor desempeño.

```python
# Celda 4
import matplotlib.pyplot as plt

# Definición de los parámetros iniciales
models_list = ['Capa Única', 'Dos Capas', 'MultiCapa']
metrics_test_mse_list = [test_metrics_1[1], test_metrics_2[1], test_metrics_3[1]]
metrics_test_mae_list = [test_metrics_1[2], test_metrics_2[2], test_metrics_3[2]]

# Organización de las métricas en una lista para facilitar el acceso en el bucle
metrics = [metrics_test_mse_list, metrics_test_mae_list]
metric_titles = ['MSE Test', 'MAE Test']

# Creación de la figura y los ejes
fig, axes = plt.subplots(1, 2, figsize=(12, 6))  # Ajusta el tamaño y la configuración de subplots

# Bucle para crear cada gráfico de barras
for idx, ax in enumerate(axes.flatten()):
    metric_values = metrics[idx]
    bars = ax.bar(models_list, metric_values, color='blue')

    # Resaltar la barra con el menor valor
    min_value = min(metric_values)
    for bar in bars:
        if bar.get_height() == min_value:
            bar.set_edgecolor('red')
            bar.set_linewidth(2)

        # Añadir etiquetas de texto sobre cada barra
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():.4f}',
                ha='center', va='bottom', color='black')

    # Configuración del título y etiquetas de los ejes
    ax.set_title(metric_titles[idx])
    ax.set_xlabel('Modelo')
    ax.set_ylabel('Valor')

plt.tight_layout()  # Ajusta automáticamente los subplots para que encajen en la figura
plt.show()  # Mostrar la figura

```
